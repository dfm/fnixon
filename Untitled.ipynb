```python
from __future__ import division, print_function
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

%matplotlib inline
```

```python
import seaborn as sns
```

```python
nasabudget = pd.read_csv('nasabudget.csv')
incarceration = pd.read_csv('incarceration.csv')
```

```python
with sns.axes_style("whitegrid"):
    fig, ax = plt.subplots(1,1)
    ax.plot(nasabudget.Year, nasabudget.Percent)
    ax2 = ax.twinx()
    ax2.plot(incarceration.Year, incarceration.IncarcerationRate/1000, )
    ax2.fill_between([1969,1974],[6,6], alpha=0.5)
    ax2.fill_between([1981,1989],[6,6], alpha=0.5)
    ax.set_xlim(1958,2014)
    ax2.set_xlim(1958,2014)
    ax.set_ylim(0,4.8)
    ax2.set_ylim(0,0.96)
    fig.savefig('fnixon.png', dpi=200)
```
